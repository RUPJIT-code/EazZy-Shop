```python
import pandas as pd
df=pd.read_csv('data.csv')
```

```python
df['Description'].duplicated().sum()  # shows how many duplicates exist
```

```python
df['Description'].nunique()
```

```python
df = df.drop_duplicates(subset=['Description'])
```

```python
# print(f"Rows before: {original_count}")
print(f"Rows after: {len(df)}")
print(f"Unique descriptions: {df['Description'].nunique()}")
```

```python
df.to_csv('data_unique.csv', index=False)
```
